```python
import random
import smtplib
from email_validator import validate_email, EmailNotValidError
import json

def load_credentials():
    try:
        with open("config.json", "r") as file:
            credentials = json.load(file)
            sender_email = credentials.get("EMAIL_ID")
            sender_password = credentials.get("EMAIL_PASSWORD")
            return sender_email, sender_password
    except Exception as e:
        print(f"Error loading credentials: {e}")
        return None, None

# Function to generate a 6-digit OTP
def generate_otp():
    return f"{random.randint(100000, 999999)}"

# Function to send the OTP via email
def send_otp(otp, recipient_email):
    sender_email, sender_password = load_credentials()

    if not sender_email or not sender_password:
        print("Error: Email credentials are not set.")
        return

    try:
        # Validate the recipient's email
        validate_email(recipient_email)

        # Gmail SMTP server details
        smtp_server = "smtp.gmail.com"
        smtp_port = 587

        # Email content
        subject = "Your OTP Code"
        message = f"Subject: {subject}\n\nYour OTP code is: {otp}"

        # Send the email
        with smtplib.SMTP(smtp_server, smtp_port) as server:
            server.starttls()
            server.login(sender_email, sender_password)
            server.sendmail(sender_email, recipient_email, message)
        print("OTP sent successfully.")
    except EmailNotValidError:
        print("Invalid email address.")
    except Exception as e:
        print(f"Error sending OTP: {e}")

# Function to verify the OTP
def verify_otp(sent_otp, entered_otp):
    return sent_otp == entered_otp

# User interaction
def user_interaction():
    otp = generate_otp()
    recipient_email = input("Enter your email address: ")

    try:
        send_otp(otp, recipient_email)
    except Exception:
        print("Failed to send OTP. Exiting.")
        return

    max_attempts = 3
    for attempt in range(max_attempts):
        entered_otp = input("Enter the OTP sent to your email: ")
        if verify_otp(otp, entered_otp):
            print("Access granted!")
            return
        else:
            print(f"Incorrect OTP. Attempts left: {max_attempts - attempt - 1}")

    print("Access denied. Too many incorrect attempts.")

# Run the system
user_interaction()
```
